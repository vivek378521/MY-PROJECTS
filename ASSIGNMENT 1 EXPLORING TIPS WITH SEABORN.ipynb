```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dftips = sns.load_dataset('tips')
dftips.head()
```

```python
sns.set_style('whitegrid')
```

# Data visualisation of how much money do majority people spend in a restaurant using violinplot.


```python
sns.violinplot(x = 'total_bill' , data = dftips)
plt.show()
```

From the above visualisation we can infer that most of the people spend over $10 to $20 in a restaurant.

# How much tip do people give on spending a specific amount of money.

```python
sns.lmplot(x = 'total_bill' , y = 'tip' , data = dftips , fit_reg = False , scatter = True)
plt.show()
```

From the above graph we can infer that as the total bill increases so does the tip.

# Relation between total bill, tip given and gender

```python
sns.lmplot(x = 'total_bill' , y = 'tip' , hue = 'sex' , palette = 'cubehelix' , fit_reg = False , 
           markers = ['o' , 'x'] , data = dftips)
plt.show()
```

We can ascertain from the graph that the lowest tips are given by the females and the highest tips are given by males but as the tips are given on the basis of the total bill generated this fact that the females are unwilling to pay more money cannot be really verified. 

# Relation between total bill, tip and time

```python
sns.lmplot(x = 'total_bill' , y = 'tip' , hue = 'day' , palette = 'rainbow' , fit_reg = False , data = dftips)
plt.show()
```

We can infer from the above graph that most of the tips were paid on Saturday and Sunday aas well as the billling amount was also maximum on Sunday and Saturday. Hence the weekends are more profitable than the weekdays.

# Relation between time and smoking activity

```python
sns.countplot(hue = 'time' , x = 'smoker' , data = dftips , palette = 'rainbow')
plt.show()
```

From the above visualiazation we can infer that the non smoker usually visit during dinner time and same is with the lunch time. The smokers do not want to be seen in the restaurant so the visit less often

# A plot of tip


```python
sns.distplot(dftips['tip'])
plt.show()
```

The graph is right skewed and there are two peaks. Most of the tip value lise between 1 to 4.

# Relation between gender and smoking habit

```python
sns.countplot(data = dftips , x = 'smoker' , hue = 'sex' , palette = 'rainbow')
plt.show
```

From the graph we can easily say that most of the smokers are men.

# Relation between total bill and smoking habit.

```python
sns.swarmplot(x = 'smoker', y = 'total_bill', data = dftips, palette = 'rainbow')
plt.show()
```

We can see that people who smoke are present in the whole range of money possession but it in the non smokers arena the concentration of the populetion is around 10-20. It can be concluded that the people who smoke tend to spend money in the whole calculated range.

```python
sns.lmplot(x = 'total_bill' , y = 'tip' , hue = 'smoker',row = 'sex' , col = 'day' , fit_reg = False , data = dftips , scatter = True)
plt.show()
```

We can infer that on friday females tend to not smoke, males smoke the most on Saturday, most women smoke on thursday.

# Relation between time and day

```python
sns.countplot(x = 'day' , hue = 'time' , palette = 'rainbow' , data = dftips)
plt.show()
```

On thursday people only come for lunch, on friday there are very few customers but on saturday and sunday the customers are very upcoming and only prefer to have dinner that explain the weekend.
